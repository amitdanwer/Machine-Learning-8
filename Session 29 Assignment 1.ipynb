```python
response = urllib.request.urlopen('http://php.net/')
```

```python
html = response.read()
```

```python
soup = BeautifulSoup(html,"html5lib")
```

```python
texts = soup.get_text()
```

```python
texts
```

```python
texts = texts.split()
```

```python
texts
```

```python
#texts = re.sub('[^a-zA-Z0-9 \n\.]', '', texts)
texts = re.sub('[\n\t.]', ' ', texts)
```

```python
texts = texts.lower()
texts
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
len(stop)
```

```python
word_tokens = word_tokenize(texts)
 
texts_f = [w for w in word_tokens if not w in stop]
 
texts_f = []
 
for w in word_tokens:
    if w not in stop:
        texts_f.append(w)
```

```python
texts_f
```

```python
# remove all tokens that are not alphabetic
texts_2 = [word for word in texts_f if word.isalpha()]
texts_2
```

```python
# stemming of words
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
texts_3 = [porter.stem(word) for word in texts_2]
```

```python
texts_3
```

```python
from collections import Counter
c = Counter((x for y in texts_3 for x in y.split()))
```

```python
c
```

```python
texts_4 = [x for x in c if c.get(x) > 2]
texts_4
```

```python
# Create freq dist and plot
freqdist1 = nltk.FreqDist(texts_4)
freqdist1.plot(25)
```
